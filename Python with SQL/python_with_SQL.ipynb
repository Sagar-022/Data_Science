```python
import pandas as pd
import sqlite3
```

```python
customers = pd.read_excel(r'C:\Users\sagar\Desktop\Data Science\My Works\Python with SQL\Customer.xlsx', sheet_name='Sheet1',header=0)
```

```python
orders = pd.read_excel(r'C:\Users\sagar\Desktop\Data Science\My Works\Python with SQL\Product.xlsx', sheet_name='Orders',header=0)
```

```python
db_conn = sqlite3.connect("super_store.db")
```

```python
c = db_conn.cursor()
```

```python
c.execute("CREATE TABLE IF NOT EXISTS customers(customer_id INTEGER NOT NULL, customer_name TEXT, province TEXT, region TEXT, customer_segment TEXT, PRIMARY KEY(customer_id))")
```

```python
c.execute("CREATE TABLE IF NOT EXISTS orders(order_id INTEGER NOT NULL, customer_id INTEGER NOT NULL, order_date TEXT, order_priority TEXT, order_quantity INTEGER, sales REAL, discount REAL, profit REAL, unit_price REAL, shipping_cost REAL, product_name TEXT, PRIMARY KEY(order_id), FOREIGN KEY(customer_id) REFERENCES customers(customer_id))")
```

```python
customers.to_sql('customers', db_conn, if_exists='append', index=False)
```

```python
orders.to_sql('orders', db_conn, if_exists='append', index=False)
```
